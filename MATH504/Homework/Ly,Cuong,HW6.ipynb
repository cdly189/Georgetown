### Question 1

#### Part a

```python
matrix = np.array([[3,-1,1],
                  [0,2,1],
                  [-1,1,4]])
```

```python
matrix
```

```python
w,v=eig(matrix)
print('Eigenvalues:', w)
print('Eigenvector', v)
```

```python
v = [1/2,1/2,1]
```

```python
## The dominant eigenvalue is 4 and its coresponding eigenvector is [1/2 1/2 1]^T
```

#### Part b

```python
def normalize(x):
    fac = abs(x).max()
    x_n = x / x.max()
    return fac, x_n

# Choose the initial vector x^(0) = [1 1 1]^T
x = np.array([-1,1,1])

print("The epsilon is: ")
for i in range(1,80):
    x = np.dot(matrix, x)
    lambda_1, x = normalize(x)
    error = LA.norm(x-v, 2) / LA.norm(v, 2)
    print('Iteration',i,': ')
    print('Error: ',error)
    print('Eigenvalue:', lambda_1)
    print('Eigenvector:', x)
    print('')
```
