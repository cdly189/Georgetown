### Exercise 4

```python
# Convert to billions in unit
convert_bi = lambda x: round(x / 1000000000,3)
df['ask_billions'] = df['avail_seat_km_per_week'].apply(convert_bi)
```

```python
# Drop a column
del df['avail_seat_km_per_week']
```

```python
df.head()
```

### Exercise 5

```python
df['fat_per_ask_85_99'] = df['fatalities_85_99'] / df['ask_billions']
df['fat_per_ask_00_14'] = df['fatalities_00_14'] / df['ask_billions']
```

```python
df.head()
```

```python
# The unit is number of fatalities per one billion killometers flown every week
```

### Exercise 6

```python
# Fatalities from 00 to 14
df['fat_per_ask_00_14'].mean()
```

```python
## The information is consistent with the idea that flying has become safer over time
## The mean, median of number of fatalities decreases 
```

### Exercise 7

```python
df['rank_fat_per_ask_85_99'] = df['fat_per_ask_85_99'].rank(ascending=False)
```

```python
df['rank_fat_per_ask_00_14'] = df['fat_per_ask_00_14'].rank(ascending=False)
```

```python
df.sort_values(by=['rank_fat_per_ask_85_99'],inplace=True)
```

### Exercise 8

```python
rank_correlation = df['rank_fat_per_ask_85_99'].corr(df['rank_fat_per_ask_00_14'])
```

```python
print(rank_correlation)
```

```python
# This is a low correlation, suggesting there is little predictability from the 85-99 to 00-14
```
